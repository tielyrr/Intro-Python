Question 1: Class Definition and Data Loading

Implement an __init__ method to load the dataset into a pandas DataFrame.

Question 2: Data Preprocessing Objective: Add data cleaning and normalization to WineDataAnalyzer. Tasks: Method handle_missing_values to manage missing data. Method normalize_data for Z-score normalization of numerical columns. Continuity: Utilize the class from Question 1.

Question 3: Grouping and Aggregation Analysis Objective: Perform analysis based on customer segments. Tasks: Create two functions in WineDataAnalyzer class average_alcohol_by_segment method for average alcohol content per segment. top_segments_by_flavanoids method to identify top three segments by flavanoids. Question 4: Data Visualization Objective: Implement visualization techniques for dataset insights. Tasks:Create four methods to the WineDataAnalyzer including: plot_correlation_heatmap for correlation matrix visualization. plot_pairplot_for_selected_features for 'Total_Phenols', 'Flavanoids', etc. plot_average_color_intensity to visualize average color intensity per segment. scatter_plot_alcohol_vs_proline to analyze alcohol and proline relationship. Continuity: Build upon the existing class.

```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
!pwd
!ls
winedf = pd.read_csv('Wine.csv')

class WineDataAnalyzer:
  def __init__(self, dataset):
    self.dataset = dataset
  #Question 1:
  def display_dataset_info(self):
    return winedf.describe(), winedf.info()
  def handle_missing_values(self):
    winedf.replace(' ', np.nan, inplace = True)
    return np.where(pd.isnull(winedf))
  #Question 2:
  def normalize_data(self):
    winedf_z_scaled = winedf.copy()
    for column in winedf_z_scaled.columns:
      winedf_z_scaled[column] = (winedf_z_scaled[column] - winedf_z_scaled[column].mean()) / winedf_z_scaled[column].std()
    return display(winedf_z_scaled.head(1))
  #Question 3:
  def average_alcohol_by_segment(self):
    wine_avg_segment = winedf.groupby("Customer_Segment").agg({"Alcohol":"mean"}).reset_index()
    return display(wine_avg_segment)
  def top_segments_by_flavaniod(self):
    wine_flavanoid_seg = winedf.groupby("Customer_Segment").agg({"Flavanoids":"sum"})
    return display(wine_flavanoid_seg)
  #Question 4:
  def plot_correlation_heatmap(self):
    sns.heatmap(winedf.corr())
    return plt.show()
  def plot_pairplot_for_selected_features(self):
    sns.pairplot(winedf, x_vars=["Total_Phenols", "Magnesium", "Ash", "Malic_Acid"], y_vars=["Color_Intensity"], hue = 'Customer_Segment')
    return plt.show()
  def plot_average_color_intensity(self):
    sns.barplot(winedf, x = "Customer_Segment", y= "Color_Intensity", hue = "Customer_Segment")
    return plt.show()
  def scatter_plot_alcohol_vs_proline(self):
    sns.scatterplot(winedf, x = 'Proline', y = 'Alcohol', hue = 'Customer_Segment')
    return plt.show()




b = WineDataAnalyzer(winedf)

print("---Question 1--- \n \n **Details and stats** \n \n", b.display_dataset_info())
print("---Question 2--- \n \n **Missing values** \n \n", b.handle_missing_values(), "\n \n **Z-Scale** \n \n", b.normalize_data())
print("---Question 3--- \n \n **Aggregation for mean** \n \n", b.average_alcohol_by_segment(), "\n \n **Agg for sum** \n \n", b.top_segments_by_flavaniod())
print("\n \n ---Question 4--- \n \n **Heatmap** \n \n")
print(b.plot_correlation_heatmap())
print("\n \n **Pairplot** \n \n")
print(b.plot_pairplot_for_selected_features())
print("\n \n **Bar graph** \n \n")
print(b.plot_average_color_intensity())
print("\n \n **Scatter Plot** \n \n")
print(b.scatter_plot_alcohol_vs_proline())
```

**MPG Dataset**

Question 1: Advanced Data Loading and Transformation
Task: Define MpgDataExpert class and load the 'mpg' dataset.Then Create a function to:
Transform 'horsepower' to numeric.
Add 'weight_to_horsepower_ratio' column.

Question 2: Complex Data Aggregation
Task: Create an aggregate_data function in MpgDataExpert class to group by 'cylinders' and 'origin'; calculate mean 'mpg' and median 'weight_to_horsepower_ratio'.

Question 3: Custom Data Filtering Method
Task:  Create custom_filter function in MpgDataExpert class to filter cars below the 25th percentile in 'mpg' or above the 75th percentile in 'weight'.

Question 4: Advanced Visualization
Task: Create plot_cylinders function in MpgDataExpert class to create a grid of scatter plots showing 'mpg' vs 'weight' for each 'cylinder' type (Tip: You can use FacetGrid to answer this question).

```python
import seaborn as sns
import pandas as pd
import numpy as np
mpg_dataset = sns.load_dataset('mpg')

class MpgDataExpert:
  def __init__(self, dataframe):
    self.dataframe = dataframe
  #Question 1:
  def numeric(self):
    #column already uniform numeric but here is how to do it:
    mpg_dataset['horsepower'] = pd.to_numeric(mpg_dataset['horsepower'])
    return (mpg_dataset['horsepower'].dtypes)
  def add_column(self):
    mpg_dataset["weight_to_horsepower_ratio"] = mpg_dataset["weight"] / mpg_dataset["horsepower"]
    return mpg_dataset.head()
  def handle_missing_vals(self):
    mpg_dataset.dropna(axis = 0, inplace = True)
    return len(mpg_dataset)
  #Question 2:
  def aggregate_data(self):
    agg_mpg = mpg_dataset.groupby(['cylinders', 'origin']).agg({'mpg':'mean', 'weight_to_horsepower_ratio':'median'})
    return agg_mpg.head()
  #Question 3:
  def custom_filter(self):
    w = mpg_dataset['weight'].describe().loc['75%']
    m = mpg_dataset['mpg'].describe().loc['25%']
    weight_75 = mpg_dataset[mpg_dataset['weight'] > w]
    mpg_25 = mpg_dataset[mpg_dataset['mpg'] < m]
    return weight_75, mpg_25
  def plot_cylinders(self):
    sns.set_context("notebook", font_scale = 1)
    g = sns.FacetGrid(mpg_dataset, col = "cylinders", hue = 'origin', despine = False)
    g.map(sns.scatterplot, "mpg", "weight", size = 8)
    g.add_legend()
    return plt.show()

a  = MpgDataExpert(mpg_dataset)
#Question 1:
print("---Question 1--- \n \n \n", "**Proof of numeric** \n", a.numeric(), "\n \n **Adding a column** \n", a.add_column(), "\n \n **Removed rows with missing values** \n", a.handle_missing_vals())
#Question 2:
#mpg_dataset.info() #checking to make sure the inplace worked so the aggregate_data can work properly
print("\n \n ---Question 2--- \n \n **Data aggregation**\n", a.aggregate_data())
#Question 3:
print("\n \n ---Question 3--- \n \n **Filtering**\n", a.custom_filter())
#Question 4:
print("\n \n ---Question 4--- \n \n **Visualization**\n")
print( a.plot_cylinders())
```
